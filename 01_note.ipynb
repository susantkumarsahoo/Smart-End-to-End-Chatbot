```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_classic.chains import ConversationChain
from langchain_classic.memory import ConversationBufferMemory
from langchain_classic.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_classic.schema import SystemMessage,HumanMessage                                                           
```

```python
# Load environment variables
load_dotenv()
```

```python
openai_api_key = os.getenv("OPENAI_API_KEY")
if not openai_api_key:
    raise ValueError("OPENAI_API_KEY not found. Please add it to your .env file.")
```

```python
# Initialize the OpenAI chat model
chat_model = ChatOpenAI(
    openai_api_key=openai_api_key,
    model_name="gpt-3.5-turbo",  # You can use gpt-4-turbo or gpt-5 if available
    temperature=0.7,  # Controls creativity (0 = factual, 1 = creative)
    max_tokens=500    # Optional: limits response length
)
# Define the prompt template
system_message = SystemMessage(
    content=(
        "You are a compassionate mental wellness assistant. "
        "Offer supportive and empathetic responses. "
        "Avoid giving medical diagnoses. "
        "Encourage mindfulness, self-reflection, and journaling."
    )
)
prompt = ChatPromptTemplate.from_messages([
    system_message,
    MessagesPlaceholder(variable_name="history"),
    ("user", "{input}")
])
# Setup memory for ongoing conversation
memory = ConversationBufferMemory(memory_key="history", return_messages=True)
# Initialize Conversation Chain
chatbot = ConversationChain(
    llm=chat_model,
    memory=memory,
    prompt=prompt,
    verbose=True
)

def main():
    print("ðŸŒ¿ Mental Wellness Support Assistant ðŸŒ¿")
    print("Type 'exit' to end the conversation.\n")

    while True:
        user_input = input("You: ")
        if user_input.lower() in ["exit", "quit"]:
            print("âœ¨ Take care of yourself. Goodbye!")
            break

        response = chatbot.run(input=user_input)
        print("AI:", response)
```
